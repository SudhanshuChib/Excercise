```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import time
```

The script below visits the home page for motors on ebay and captures all names and url's of car makers categories (Honda, Toyota, Tesla etc.).

Based on the list of car makers individual categories are scrapped and saved in a CSV file.

```python

HomeUrl="https://www.ebay.com/sch/Cars-Trucks/6001/i.html?LH_ItemCondition=2%7C0&_dcat=6001&_dmpt=US_Cars_Trucks"
r=requests.get(HomeUrl)
soup=BeautifulSoup(r.content,"html.parser")
#Finding Brand names
Brands=soup.find_all("div",{"class":"cat-link"})
links=[]
breakPt=len(Brands)
for i in range(breakPt):
    j=i+1
    if j == breakPt:
        break
    else:
        WebAddress = re.search(r'href=[\'"]?([^\'" >]+)', str(Brands[j]))
        WebAddress = WebAddress.group(0)
        links.append(WebAddress[6:])

	
#Declaring lists to store values
carName=[]
carPrice=[]
carYear=[]
carMileage=[]

#Importing following pages
#Setting Variable for eBay pages that will regulate how many calls are send to the website
#Forming URL

for k in range(len(Brands)-2):
	Url1=links[k]
	Page=2
	Count=50
	CarsFound="Yes"
	print(Brands[k].text)
	
	#Importing URL
	r=requests.get(Url1)
	soup=BeautifulSoup(r.content,"html.parser")

	Name=soup.find_all("a",{"class":"vip"})
	Price=soup.find_all("li",{"class":"lvprice prc"})
	Year=soup.find_all("ul",{"class":"lvdetails left space-zero full-width"})
	
	if not Name:
		CarsFound="No"
		continue
	else:
		for i in range(len(Name)):
			#Adding Car Name
			try:  
				carName.append(Name[i].text.replace("\n","").replace("\t","").replace("\r","").replace("New listing",""))
			except:
				carName.append("NotFound")

			#Adding Car Price
			try:
				carPrice.append(Price[i].text.replace("\n","").replace("\t","").replace("$","").replace(",",""))
			except:
				carPrice.append("NotFound")

			#Adding Car Year
			try:
				year=str(Year[i].text.replace("\n","").replace("\t",""))
				year=(re.search(r'Year: \d\d\d\d',year))
				year=year.group(0)
				carYear.append(year.replace("Year: ",""))
			except:
				carYear.append("NotFound")

			#Adding Car Mileage
			try:
				mileage=str(Year[i].text.replace("\n","").replace("\t","").replace(",",""))
				mileage=(re.search(r'Mileage: \d+',mileage))
				mileage=mileage.group(0)
				carMileage.append(mileage.replace("Mileage: ",""))
			except:
				carMileage.append("NotFound")            

	while CarsFound == 'Yes':
	#for a in range(3): 
		FinalURL=Url1+"&_pgn="+str(Page)+"&_skc="+str(Count)
		Page = Page + 1
		Count = Count + 50
		time.sleep(5)
		#print(FinalURL)
		if Page%5==0:
			print(Page)
		#Importing URL
		r=requests.get(FinalURL)
		soup=BeautifulSoup(r.content,"html.parser")

		Name=soup.find_all("a",{"class":"vip"})
		Price=soup.find_all("li",{"class":"lvprice prc"})
		Year=soup.find_all("ul",{"class":"lvdetails left space-zero full-width"})
		
		if not Name:
			CarsFound="No"
		else:
			for i in range(len(Name)):
				#Adding Car Name
				try:  
					carName.append(Name[i].text.replace("\n","").replace("\t","").replace("\r","").replace("New listing",""))
				except:
					carName.append("NotFound")

				#Adding Car Price
				try:
					carPrice.append(Price[i].text.replace("\n","").replace("\t","").replace("$","").replace(",",""))
				except:
					carPrice.append("NotFound")

				#Adding Car Year
				try:
					year=str(Year[i].text.replace("\n","").replace("\t",""))
					year=(re.search(r'Year: \d\d\d\d',year))
					year=year.group(0)
					carYear.append(year.replace("Year: ",""))
				except:
					carYear.append("NotFound")

				#Adding Car Mileage
				try:
					mileage=str(Year[i].text.replace("\n","").replace("\t","").replace(",",""))
					mileage=(re.search(r'Mileage: \d+',mileage))
					mileage=mileage.group(0)
					carMileage.append(mileage.replace("Mileage: ",""))
				except:
					carMileage.append("NotFound")            
```

```python
ebay= pd.DataFrame(
    {"Name": carName,
     "Year":carYear,
     "Price":carPrice,
     "Mileage":carMileage
    })

#Saving a copy on hard disk
ebay.to_csv('C:\\Users\\sudhanshu\\Desktop\\Springboard\\Used Cars Dataset\\ebayRevised.csv', index=True, header=True)
```
